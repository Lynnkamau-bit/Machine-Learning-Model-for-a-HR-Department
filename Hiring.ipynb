```python
import pandas as pd
import numpy as np
import csv
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/Python files/hiring (1).csv')
df
```

```python
#Data pre-processing: Fill NA values with the median value of a column
df['test_score(out of 10)'].median()
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
```

```python

df['experience'] = df['experience'].replace({
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11
})
df['experience'].median()
df['experience'] = df['experience'].fillna(df['experience'].median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('salary($)', axis = 'columns'), df['salary($)'])
```

```python
reg.intercept_
```

```python
reg.score
```
